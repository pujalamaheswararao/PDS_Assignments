```python
data['Total_Score'] = data['math score'] + data['writing score'] + data['reading score']
```

```python
import matplotlib.pyplot as plt
```

```python
gender_counts = data['gender'].value_counts()
plt.bar(gender_counts.index, gender_counts.values)
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.savefig ('E:\\maheshswararo pujala\\Assignment1\\ASS1.Question2\\results\\Gender_distribution.png')
plt.show()
```

```python
sns.boxplot(x='gender', y='average', data=data)
plt.title('Average by Gender')
plt.savefig ('E:\\maheshswararo pujala\\Assignment1\\ASS1.Question2\\results\\Math_Scores_by_Gender.png')
plt.show()
```

```python
sns.kdeplot(x='reading score', data=data, hue='lunch', fill=True, alpha=.3)
plt.title('Reading Scores by Lunch Type')
plt.xlabel('Reading Score')
plt.savefig ('E:\\maheshswararo pujala\\Assignment1\\ASS1.Question2\\results\\Reading_Scores_by_Lunch_Type.png')
plt.show()
```
